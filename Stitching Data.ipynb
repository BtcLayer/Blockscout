```python
df = pd.DataFrame()
sym="UNIUSDT"
y="2023"
for i in range(1, 7):
    mon=i
    if(mon<10):
        st="0"+str(mon)
    else:
        st=str(mon)
    tp = pd.read_csv('Raw Data/'+sym+'/'+y+'/'+sym+'-1m-'+y+'-'+st+'.csv', names=column_names)
    df=pd.concat([df, tp])
    df.reset_index(drop=True, inplace=True)
df
```
